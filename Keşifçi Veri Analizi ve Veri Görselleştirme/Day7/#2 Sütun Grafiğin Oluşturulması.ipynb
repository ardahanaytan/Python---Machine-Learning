```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

```python
df['cut'].value_counts().plot.barh();
```

```python
df['cut'].value_counts().plot.barh().set_title('Cut Degiskeninin Sinif Frekanslari');
```

```python
sns.barplot(x = 'cut', y = df.cut.index, data = df);
```
